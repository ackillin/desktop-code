```python
from qiskit import *
from qiskit.visualization import plot_histogram
from qiskit_ibm_provider import IBMProvider
from qiskit.circuit import Bit
from numpy import pi
```

```python
provider = IBMProvider()
```

```python
backend = provider.get_backend('ibm_nairobi')  # Simulator for now
```

# Making it work on actual quantum computer
Gates allowed: CX, ID, RZ, SX, X

```python
num_qubits = 3
qc = QuantumCircuit(num_qubits)
```

```python
[qc.sx(i) for i in range(num_qubits) if i%2 == 0]
```

```python
qc.cx(0,1)
#qc.cx(1,2)
#qc.cx(2,3)
#qc.cx(4,5)
#qc.cx(6,7)
```

```python
qc.measure_all()
```

```python
qc.draw('mpl')
```

```python
job = backend.run(qc, shots = 1000)
```

```python
results = job.result()
counts = results.get_counts()
```

```python
counts
```

```python
plot_histogram(counts)
```
